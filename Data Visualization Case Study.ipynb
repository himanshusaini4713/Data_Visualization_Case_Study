### Q6. Create a new column Qtr using numpy.where() or any suitable utility in the imported dataset. The Quarters are based on months and defined as -
#### • Jan-Mar:Q1 
#### • Apr-Jun:Q2 
#### • Jul-Sep:Q3 
#### • Oct-Dec:Q4

```python
#Through np.where 
```

```python
Data["Qtr1"] = np.where(Data.Month.isin(["Jan","Feb","Mar"]),"Q1",np.where(Data.Month.isin(["Apr","May","Jun"]),"Q2",np.where(Data.Month.isin(["Jul","Aug","Sep"]),"Q3","Q4")))
```

```python
Data
```

```python
#Through If Else
```

```python
month=Data['Month']
```

```python
quater = []
for x in month :
    if x in ["Jan","Feb","Mar"]:
        quater.append("Q1")
    elif x in ["Apr","May","Jun"]:
        quater.append("Q2")
    elif x in ["Jul","Aug","Sep"]:
        quater.append("Q3")
    else:
        quater.append("Q4")
```

```python
quater

```

```python
Data["Qtr"]= pd.Series(quater)
```
